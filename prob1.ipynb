```python
//Question 1: 
//Visualizing Multidimensional Random Data using Scatter Plot
//Create a scatter plot with randomly generated data points that vary in color, size, and transparency to represent multiple dimensions of information within a single visualization
//Task:
//Write a Python program to:
//Generate 100 random data points for:
//x and y coordinates using a normal distribution
//random color values between 0 and 1
//random sizes (scaled up for better visibility)
//Create a scatter plot using matplotlib.pyplot.scatter() with:
//c=colors → to assign different colors to each point
//s=sizes → to vary the size of each point
//alpha=0.3 → to set transparency for overlapping points
//cmap='viridis' → to apply a smooth color gradient
//Display a color scale using plt.colorbar().
//Label the axes and give the plot a suitable title such as “Random Multidimensional Scatter Plot”.
//Expected Output:
//A colorful scatter plot showing 100 data points of varying colors and sizes, with transparency and a visible color scale. The plot should visually demonstrate how color, size, and transparency can represent multiple variables simultaneously.

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
x = np.random.randn(100)         # x coordinates (normal distribution)
y = np.random.randn(100)         # y coordinates (normal distribution)
colors = np.random.rand(100)     # random color values between 0 and 1
sizes = 1000 * np.random.rand(100)  # random sizes (scaled up)

# Create scatter plot
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')

# Add color bar
plt.colorbar(label='Color Scale')

# Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Multidimensional Scatter Plot')

# Show plot
plt.show()
```
