```python
import numpy as np
import matplotlib.pyplot as plt

# Generate first dataset (Group A)
x1 = np.random.normal(loc=0, scale=1, size=100)
y1 = np.random.normal(loc=0, scale=1, size=100)

# Generate second dataset (Group B)
x2 = np.random.normal(loc=4, scale=1.5, size=100)
y2 = np.random.normal(loc=4, scale=1.5, size=100)

# Create scatter plot
plt.scatter(x1, y1, color='blue', label='Group A', alpha=0.7)
plt.scatter(x2, y2, color='red', label='Group B', alpha=0.7)

# Add title and labels
plt.title("Scatter Plot of Two Normal Distributions")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add legend
plt.legend()

# Add grid for better visualization
plt.grid(True)

# Display the plot
plt.show()
```
